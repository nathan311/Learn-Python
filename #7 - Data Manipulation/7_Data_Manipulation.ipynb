```
df_france_death = pd.DataFrame(time_series_df_deaths).reset_index()
df_france_death.columns = ['date', 'cases_number']
df_france_death
```

```
df_france_recovered = pd.DataFrame(time_series_france).reset_index()
df_france_recovered.columns = ['date', 'cases_number']
df_france_recovered
```

```
df_france_confirmed = pd.DataFrame(time_series_france_confirmed).reset_index()
df_france_confirmed.columns = ['date', 'cases_number_confirmed']
df_france_confirmed
```

```
df_france = df_france_death.merge(df_france_recovered, on='date')
df_france
```

```
df_france = df_france.merge(df_france_confirmed, on='date')
df_france
```

```
df_france.columns = ['date','cases_number_deaths','cases_number_recovered','cases_number_confirmed']
df_france
```
