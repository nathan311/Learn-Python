```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")

plt.figure(figsize=(5,5))


sns.pairplot(df[['calories', 'protein','fat', 'sugars', 'vitamins', 'shelf']], hue="shelf")
```

```python
sns.countplot(df, x="fat", hue="type")
```

```python
import plotly.express as px
fig = px.scatter(df, x="sugars", y="rating", color = 'shelf')
fig.show()
```
